```python
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)
import matplotlib.pyplot as plt

```

```python
df= sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

## Data Visualisation

## Distribution Plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['total_bill'],color='y')
```

```python
sns.distplot(df['size'],color='g')
```

##Pair Plot

## Pair Plot

```python
sns.pairplot(df)
```

## Pair plot Of Separated Gender For this data Set

```python
sns.pairplot(df,hue='sex',palette='husl')
```

```python
sns.pairplot(df,hue='smoker',palette='bright')
```

```python
sns.pairplot(df,hue='day',palette='dark')
```

## Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=df,palette='colorblind')
```

## Barplot between Smokers and gender

```python
sns.barplot(y='smoker',x='tip',data=df,palette='dark')
```

## Day V/S Tip

```python
sns.barplot(x='day',y='tip',hue='smoker',data=df,palette='magma')
```

```python
sns.barplot(x='time',y='total_bill',hue='sex',data=df,palette='pastel')
```

## Count Plot

```python
# count plot for categorical data
df['sex'].value_counts()
```

## count plot between male and female

```python
sns.countplot('sex',data=df,palette='muted')
```

## Smoker V/S Non Smokers

```python
sns.countplot('smoker',data=df,palette='colorblind')
```

```python
df['day'].value_counts()
```

## Countplot Of Days

```python
sns.countplot('day',data=df,palette='hls')
```

## Smokers Visiting on particular day of week

```python
sns.countplot('day',hue='smoker',data=df,palette='dark')
```

## Violin Plots

## To find the total bill paid by each sex

```python
sns.violinplot(x='sex',y='total_bill',data=df,palette='rocket')
```

## PLOT between time,bill and smokers

```python
sns.violinplot(x='total_bill',y='time',data=df,hue='smoker',palette='Accent')
```

## BOX PLOT

```python
## Box plot for numeric data between smokers and size
sns.boxplot('smoker','size',data=df,palette='husl')
```

## Joint Plot

## plot between total_bill and tip

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

## PLot Between total_bill and size

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg',color='r')
```

## Scatter Plots

## scatter plot between day and tip recieved

```python
sns.catplot(x='day',y='tip',data=df)
```

## Scatter plot to find tip given by each sex

```python
sns.catplot(x='sex',y='tip',data=df,jitter=False)
```

## Scatter plot between smoker, total_bill, sex

```python
sns.catplot(x='smoker',y='total_bill',data=df,kind='swarm',hue='sex')
```

## Heat Maps

```python
sns.heatmap(df.corr(),annot=True)
```
